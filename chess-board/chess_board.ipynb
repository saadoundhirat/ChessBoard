```python
class ChessBoard:
    def __init__(self):
        self.grid = np.zeros((8,8,4) , np.float64)
        self.red_row =''
        self.red_col =''
        self.blue_row =''
        self.blue_col =''

        for i in range(8):
            for j in range(8):
                if (i + j) % 2 == 0:
                    self.grid[i,j] = [1,1,1,0.7]
                else:
                    self.grid[i,j] = [0,0,0,0.7]

    def render(self):
        """adding render method to show the image"""
        plt.imshow(self.grid)

    def red_queen(self , row:int ,col:int):
        """adding red queen"""
        self.red_row =row
        self.red_col =col
        self.grid[row,col] = [1,0,0,0.2]
        self.render()
        return self.grid

    def blue_queen(self , row:int ,col:int):
        """adding blue queen"""
        self.blue_row =row
        self.blue_col =col
        self.grid[row,col] = [0,0,1,0.2]
        self.render()
        return self.grid

    def queen_under_attack(self):
        """To see if the queen is under attack"""
        # if self.add_blue_queen ==
        if self.red_row == self.blue_row  or self.red_col == self.blue_col:
            return True
        if abs(self.blue_row - self.red_row) == abs(self.blue_col - self.red_col):
            return True
        return False

```

```python
def test_linear_attack():
    chessboard = ChessBoard()
    chessboard.red_queen(0,0)        
    chessboard.blue_queen(7,7)        
    chessboard.render()
    assert chessboard.queen_under_attack() == True
    return chessboard.queen_under_attack()
test_linear_attack()
```
